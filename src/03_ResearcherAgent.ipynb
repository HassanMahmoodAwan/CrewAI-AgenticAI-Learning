```python
from crewai import LLM, Agent, Task, Crew

ai_model = LLM(
    model = "gpt-4.1-mini",
    temperature = 0.1,
)



researcherAssistant = Agent(
    role="Reasearch specialist",
    goal="Search deeply and in detail about the topic: {topic} and all its interesting facts",
    backstory="You are expert and skilled in deep and accurate reasearch about the {topic}. You know how to search expertly",
    llm=ai_model,
    tools=[],
    verbose=False
    
) 

writterAssistant = Agent(
    role="Creative Writer",
    goal="Write a short summary Blog about the topic in intresting and engaging way.",
    backstory="you are expert and skilled in writing Blogs in summarized way. you will be provided with the research about the topic: {topic}, you will writed a summarized blog for it.",
    llm=ai_model,
    tools=[],
    verbose=False
    
) 



researchTask = Task(
    description="Reasearch about all the interesting facts and information about {topic}",
    expected_output="Provided search in paragraph, but facts or interesting points will be in bullet points.",
    agent=researcherAssistant
)

writerTask = Task(
    description="Write a summary blog about the topic {topic}",
    expected_output="Provide summary in short paragraph with points in bullet form",
    agent=writterAssistant,
    context = [researchTask]
)



crew = Crew(
    agents=[researcherAssistant, writterAssistant],
    tasks=[researchTask, writerTask],
    verbose=True
    
)


result = crew.kickoff(inputs={"topic": "Use of EVs vehical in Present & Future."})
```
